```python
%matplotlib inline 
#para dibujar en el propio notebook
import pandas as pd # se importa pandas como pd
import numpy as np  #numpy como np
import matplotlib.pyplot as plt #matplotlib.pyplot como plot
import pickle as pk
import loadData as ld
from transformers import Normalizer, NoiseFilter, StatisticsTransformer
```

```python
with open('datos.pdd','rb') as f: 
    datos = pk.load(f) 

print(len(datos))
datos.head()
```

```python
with open('datosEstadisticos.pdd','rb') as f: 
    datosEstadisticos = pk.load(f) 

print(len(datosEstadisticos))
datosEstadisticos.head()
```

## Filtro de ruido: 

```python
noiseFilter = NoiseFilter()
datos = noiseFilter.fit_transform(datos, 5)
datos.head()
```

```python
presiones = pd.DataFrame()

for i in range(1,13): 
    presiones = pd.concat([presiones, datos['P'+str(i)]], axis = 1)
presiones.head()
```

## Filtro de normalización: 

```python
normalizer = Normalizer()
datosNormalizados = normalizer.fit_transform(presiones)
```

```python
datosNormalizados.head()
```

## Transformadores estadísticos 
### - Media: 

```python
statisticTransformer = StatisticsTransformer()
mediasMoviles = statisticTransformer.fit_transform(datosNormalizados, mode='mean', window=4)
mediasMoviles.head()
```

### - Desviación estándar: 

```python
desviacionesMoviles = statisticTransformer.fit_transform(datosNormalizados, mode='std', window=4)
desviacionesMoviles.head()
```

### - Rangos: 

```python
rangosMoviles = statisticTransformer.fit_transform(datosNormalizados, mode='range', window=4)
rangosMoviles.head()
```
