# Ejemplos de uso de los transformadores: 

## Filtro de ruido: 

```python
noiseFilter = NoiseFilter(5)
datos = noiseFilter.fit_transform(datos)
datos.head()
```

```python
presiones = pd.DataFrame()

for i in range(1,13): 
    presiones = pd.concat([presiones, datos['P'+str(i)]], axis = 1)
presiones.head()
```

## Filtro de normalización: 

```python
normalizer = Normalizer(100)
datosNormalizados = normalizer.fit_transform(presiones)
datosNormalizados.head()
```

## Transformador VarianceThresholdPD: 

```python
varianceThresholdPD = VarianceThresholdPD(0.5)
datosNormalizados = varianceThresholdPD.fit_transform(datosNormalizados)
datosNormalizados.head()
```

## Transformadores estadísticos 
### - Media: 

```python
statisticTransformer = StatisticsTransformer(mode='mean', window=4)
mediasMoviles = statisticTransformer.fit_transform(datosNormalizados)
mediasMoviles.head()
```

### - Desviación estándar: 

```python
statisticTransformer2 = StatisticsTransformer(mode='std', window=4)
desviacionesMoviles = statisticTransformer2.fit_transform(datosNormalizados)
desviacionesMoviles.head()
```

### - Rangos: 

```python
statisticTransformer3 = StatisticsTransformer(mode='range', window=4)
rangosMoviles = statisticTransformer3.fit_transform(datosNormalizados)
rangosMoviles.head()
```

## Filtro signal.butter

```python
butterTransformer = ButterTransformer(N=3, Wn=0.05)
senalFiltrada = butterTransformer.fit_transform(datosNormalizados)
senalFiltrada.head()
```

## Filtro sg.savgol_filter

```python
savgolTransformer = SavgolTransformer(25,polyorder=2)
senalFiltrada = savgolTransformer.fit_transform(datosNormalizados)
senalFiltrada.head()
```
