```python
%matplotlib inline 
#para dibujar en el propio notebook
import pandas as pd # se importa pandas como pd
import numpy as np  #numpy como np
import matplotlib.pyplot as plt #matplotlib.pyplot como plot
import pickle as pk
import loadData as ld
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import metrics as mtr
import seaborn as sns
import tsfresh as tf
```

```python
features_diacrisis1_p1 = pk.load(open('features_diacrisis1_p1.pdd','rb'))
features_diacrisis1_p2 = pk.load(open('features_diacrisis1_p2.pdd','rb'))
features_diacrisis1 = pd.concat([features_diacrisis1_p1, features_diacrisis1_p2], axis = 0, ignore_index = True) 
print('Filas: ',features_diacrisis1.shape[0])
print('Número de características (crisis 1):',features_diacrisis1.shape[1]-1)
features_diacrisis1.head()
```

```python
features_diacrisis2_p1 = pk.load(open('features_diacrisis2_p1.pdd','rb'))
features_diacrisis2_p2 = pk.load(open('features_diacrisis2_p2.pdd','rb'))
features_diacrisis2 = pd.concat([features_diacrisis2_p1, features_diacrisis2_p2], axis = 0, ignore_index = True)   
print('Filas: ',features_diacrisis2.shape[0])
print('Número de características (crisis 2):',features_diacrisis2.shape[1]-1)
features_diacrisis2.head()
```

### Eliminar columnas con NaN (no se admiten en la selección de características): 

```python
#elimino características con NaN
features = pd.concat([features_diacrisis1, features_diacrisis2], axis = 0, ignore_index = True) 
features = features.dropna(1)
features_diacrisis1 = features[:121889]
features_diacrisis2 = features[121889:]
print('Filas (crisis 1):',features_diacrisis1.shape[0])
print('Filas (crisis 2):',features_diacrisis2.shape[0])
print('Número de características tras quitar nulos:',features.shape[1]-1)
features.head()
```

```python
#calculamos las características que se mantienen para todas las presiones tras eliminar los nulos 
caracteristicas = {}
for c in features: 
    if c[2:] in caracteristicas: 
        caracteristicas[c[2:]]+=1
    else: 
        caracteristicas[c[2:]]=1
```

```python
#elimino las características que no existen para todas las presiones 
columnas = features.columns
for k,v in caracteristicas.items(): 
    if v < 6: 
        for i in range(1,7): 
            nombre = "P"+str(i)+k
            if nombre in columnas: 
                features = features.drop([nombre],axis=1)
                features_diacrisis1 = features_diacrisis1.drop([nombre],axis=1)
                features_diacrisis2 = features_diacrisis2.drop([nombre],axis=1)
len(features.columns)
```

### Selección de características: 

```python
selection = tf.feature_selection.selection.select_features(features.drop(['target'],axis=1),features['target'],ml_task='classification')
selection = pd.concat([selection,features['target']],axis=1)
selection_diacrisis1 = selection[:121889]
selection_diacrisis2 = selection[121889:]
print('Número de características tras selección (crisis 1):',selection_diacrisis1.shape[1]-1)
print('Número de características tras selección (crisis 2):',selection_diacrisis2.shape[1]-1)
```

```python
#calculamos las características que se mantienen para todas las presiones tras hacer la selección
caracteristicas = {}
for c in selection.columns: 
    if c[2:] in caracteristicas: 
        caracteristicas[c[2:]]+=1
    else: 
        caracteristicas[c[2:]]=1
caracteristicas.values()
```

```python
#elimino las características que no existen para todas las presiones 
columnas = selection_diacrisis1.columns
for k,v in caracteristicas.items(): 
    if v < 6: 
        for i in range(1,7): 
            nombre = "P"+str(i)+k
            if nombre in columnas: 
                selection_diacrisis1 = selection_diacrisis1.drop([nombre],axis=1)
                selection_diacrisis2 = selection_diacrisis2.drop([nombre],axis=1)
len(selection_diacrisis1.columns)
```

```python
with open('selection_diacrisis1.pdd','wb') as f:
    pk.dump(selection_diacrisis1,f)
with open('selection_diacrisis2.pdd','wb') as f: 
    pk.dump(selection_diacrisis2,f)
```

```python
with open('selection_diacrisis1.pdd','rb') as f: 
    selection_diacrisis1 = pk.load(f) 
with open('selection_diacrisis2.pdd','rb') as f: 
    selection_diacrisis2 = pk.load(f) 
```

```python
selection_diacrisis1 = selection_diacrisis1.drop(['P1__sample_entropy','P2__sample_entropy','P3__sample_entropy','P4__sample_entropy','P5__sample_entropy','P6__sample_entropy'], axis=1)
selection_diacrisis2 = selection_diacrisis2.drop(['P1__sample_entropy','P2__sample_entropy','P3__sample_entropy','P4__sample_entropy','P5__sample_entropy','P6__sample_entropy'], axis=1)
```

```python
resultados_roc = {}
resultados_prc = {}
columnas_usadas = set()
rfc = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
y1 = selection_diacrisis1['target']
y2 = selection_diacrisis2['target']

for column in selection_diacrisis1.columns: 
    if column!='target':
        column_name = column[2:]
        if column_name not in columnas_usadas: 
            print(column_name)
            columnas_usadas.add(column_name)
            X1 = pd.DataFrame()
            X2 = pd.DataFrame()
            for P in ['P1','P2','P3','P4','P5','P6']: 
                P_column_name = P+column_name
                X1 = pd.concat([X1, selection_diacrisis1[P_column_name]],axis=1)
                X2 = pd.concat([X2, selection_diacrisis2[P_column_name]],axis=1)
                
            rfc.fit(X1, y1)
            y_pred = rfc.predict_proba(X2)
            y_score = [i[0] for i in y_pred]
            roc1 = mtr.roc_auc_score(y2,y_score)
            prc1 = mtr.average_precision_score(y2,y_score)
            
            rfc.fit(X2, y2)
            y_pred = rfc.predict_proba(X1)
            y_score = [i[0] for i in y_pred]
            roc2 = mtr.roc_auc_score(y1,y_score)
            prc2 = mtr.average_precision_score(y1,y_score)
            
            roc= (roc1+roc2)/2
            resultados_roc[column_name] = roc
            prc= (prc1+prc2)/2
            resultados_prc[column_name] = prc
```

```python
#guardar resultados
with open('roc.pkl','wb') as f:
    pk.dump(resultados_roc,f)
with open('prc.pkl','wb') as f:
    pk.dump(resultados_prc,f)
```

```python
#cargar resultados
resultados_roc = pk.load(open('roc.pkl','rb'))
resultados_prc = pk.load(open('prc.pkl','rb'))
```

### Calcular ranking para roc: 

```python
roc_ranking = pd.DataFrame()
roc_ranking['Feature'] = resultados_roc.keys()
roc_ranking['ROC'] = resultados_roc.values()
roc_ranking = roc_ranking.sort_values(by='ROC', ascending=False).reset_index(drop=True)
roc_ranking.head(10)
```

### Calcular ranking para prc: 

```python
prc_ranking = pd.DataFrame()
prc_ranking['Feature'] = resultados_prc.keys()
prc_ranking['PRC'] = resultados_prc.values()
prc_ranking = prc_ranking.sort_values(by='PRC', ascending=False).reset_index(drop=True)
prc_ranking.head(10)
```

### Añadir características al entrenamiento del clasificador hasta que la roc deje de mejorar: 

Combinación 0 -> característica 0

Combinación 1 -> características 0 y 1

Combinación 2 -> características 0,1 y 2

...

```python
def combinar_1(selection_diacrisis1, selection_diacrisis2, ranking=roc_ranking, metrica=mtr.roc_auc_score): 
    
    resultados={}
    rfc = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
    X1 = pd.DataFrame()
    X2 = pd.DataFrame()
    y1 = selection_diacrisis1['target']
    y2 = selection_diacrisis2['target']
    for i, row in ranking.iterrows(): 
        for j in range(1,7): 
            X1 = pd.concat([X1, selection_diacrisis1['P'+str(j)+row.Feature]], axis=1)
            X2 = pd.concat([X2, selection_diacrisis2['P'+str(j)+row.Feature]], axis=1)

        rfc.fit(X1, y1)
        y_pred = rfc.predict_proba(X2)
        y_score = [i[0] for i in y_pred]
        auc1 = metrica(y2,y_score)

        rfc.fit(X2, y2)
        y_pred = rfc.predict_proba(X1)
        y_score = [i[0] for i in y_pred]
        auc2 = metrica(y1,y_score)

        auc= (auc1+auc2)/2
        
        print('combinacion:',i, str(metrica)[10:-19], auc)
        if i!=0 and auc < resultados[i-1]:
            print("Deja de mejorar")
            break
        resultados[i] = auc
        
    return resultados 
```

```python
resultados2 = combinar_1(selection_diacrisis1, selection_diacrisis2, ranking=roc_ranking, metrica=mtr.roc_auc_score)
resultados2
```

```python
resultados2 = combinar_1(selection_diacrisis1, selection_diacrisis2, ranking=prc_ranking, metrica=mtr.average_precision_score)
resultados2
```

### Combinar la mejor característica con el resto para ver si el roc mejora: 

Combinación 1 -> características 0 y 1

Combinación 2 -> características 0 y 2

Combinación 3 -> características 0 y 3

...

```python
def reset_X(): 
    X1 = pd.DataFrame()
    X2 = pd.DataFrame()
    for j in range(1,7): 
        X1 = pd.concat([X1, selection_diacrisis1['P'+str(j)+roc_ranking.Feature[0]]], axis=1)
        X2 = pd.concat([X2, selection_diacrisis2['P'+str(j)+roc_ranking.Feature[0]]], axis=1)
    return X1,X2

def combinar_2(selection_diacrisis1, selection_diacrisis2, ranking=roc_ranking, metrica=mtr.roc_auc_score): 
    resultados = {}
    rfc = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
    y1 = selection_diacrisis1['target']
    y2 = selection_diacrisis2['target']
    for i, row in ranking.iterrows():
        X1,X2 = reset_X()
        if i!=0: 
            for j in range(1,7): 
                X1 = pd.concat([X1, selection_diacrisis1['P'+str(j)+row.Feature]], axis=1)
                X2 = pd.concat([X2, selection_diacrisis2['P'+str(j)+row.Feature]], axis=1)

            rfc.fit(X1, y1)
            y_pred = rfc.predict_proba(X2)
            y_score = [i[0] for i in y_pred]
            auc1 = metrica(y2,y_score)

            rfc.fit(X2, y2)
            y_pred = rfc.predict_proba(X1)
            y_score = [i[0] for i in y_pred]
            auc2 = metrica(y1,y_score)

            auc= (auc1+auc2)/2
            print('combinacion:',i, str(metrica)[10:-19], auc)
            if auc > ranking.iloc[0,1]: 
                resultados[i] = auc
    return resultados
```

```python
resultados3 = combinar_2(selection_diacrisis1, selection_diacrisis2, ranking=roc_ranking, metrica=mtr.roc_auc_score)
resultados3
```

```python
resultados3 = combinar_2(selection_diacrisis1, selection_diacrisis2, ranking=prc_ranking, metrica=mtr.average_precision_score)
resultados3
```
